```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Replace 'data.csv' with the actual path to your CSV file
df = pd.read_csv(r'Y:\LP_birds\LP_birds_2023\ARUs\ARU_audiofiles\aru_all_detections.csv')
print(df.head())
print(len(df))
```

```python
# get dataset summary values

# Calculate the average confidence and standard deviation
average_confidence = df['confidence'].mean()
std_deviation_confidence = df['confidence'].std()

# Calculate the total number of unique common names
unique_common_names = df['common_name'].nunique()

# Find the date range
date_range = df['date'].agg(['min', 'max'])

# Find the time when the greatest number of calls happen
most_common_time = df['time'].mode().iloc[0]

# Print the summary information
print("Average Confidence:", average_confidence)
print("Standard Deviation of Confidence:", std_deviation_confidence)
print("Total Number of Unique Common Names:", unique_common_names)
print("Date Range:", date_range)
print("Time with the Greatest Number of Calls:", most_common_time)
```

```python
# Example for 'confidence' column
sns.histplot(df['confidence'], bins=20)
plt.title('Histogram of Confidence')
plt.xlabel('Confidence')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count the common name occurrences and sort by frequency
sorted_df = df["common_name"].value_counts().reset_index()
sorted_df.columns = ["common_name", "Frequency"]


# Use Seaborn to create a sorted bar plot
plt.figure(figsize=(20, 15))  # Adjust the figure size as needed
sns.barplot(x="Frequency", y="common_name", data=sorted_df)
plt.xlabel('Call Frequency')
plt.ylabel('Common Name')
plt.title('Common Name Detection Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Count the common name occurrences and sort by frequency
sorted_df = df["common_name"].value_counts().reset_index()
sorted_df.columns = ["common_name", "Frequency"]


# Use Seaborn to create a sorted bar plot
plt.figure(figsize=(20, 10))  # Adjust the figure size as needed
sns.barplot(x="Frequency", y="common_name", data=sorted_df[sorted_df["Frequency"] > 100])
plt.xlabel('Call Frequency')
plt.ylabel('Common Name')
plt.title('Bird Call Detection Frequency (August 7-14, 2023)')

# Save the figure as an image file (e.g., PNG)
plt.savefig(r'Y:\LP_birds\LP_birds_2023\ARUs\Figs\call_detection_frequency.png')

# Show the plot
plt.tight_layout()
plt.show()
```
